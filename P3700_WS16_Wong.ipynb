```python
#i
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

img1=mpimg.imread("clock1.png")
img2=mpimg.imread("clock2.png")
#should have same limits
col = len(img1)
row = len(img1[0])
#output empty 2d array
imgsub=[[0 for x in range(row)] for y in range(col)] 
n = 0
while n < col:
    i = 0
    while i < row:
        imgsub[n][i]=(img1[n][i]-img2[n][i] + 1)*0.5 #scale image between 0 and 1
        i+=1
    n+=1
flatten=np.array(imgsub)
flatten=flatten.ravel()
plt.hist(flatten)
plt.show()
# this writes out the 2D list imgsub into a png file
mpimg.imsave("clocksub.png",imgsub)
```

```python
#2 filter kernals 
#a) create filter kernals
low_pass = [[1/9,1/9,1/9],[1/9,1/9,1/9],[1/9,1/9,1/9]]
high_pass = [[0,-1,0],[-1,4,-1],[0,-1,0]]
ver_sob = [[-1,0,0],[-2,0,2],[-1,0,1]]
hor_sob = [[-1,-2,-1],[0,0,0],[1,2,1]]
k = 2
edge=[[-k/8,-k/8,-k/8],[-k/8,k+1,-k/8],[-k/8,-k/8,-k/8]]
shift=[[0,0,0],[0,1,0],[0,0,-1]]
```

```python
#test on base image
import scipy.misc
import time
ascent = scipy.misc.ascent()
plt.gray()
plt.imshow(ascent)
plt.show()
```

```python
from scipy import signal
time_start=time.clock()
output=signal.convolve2d(low_pass,ascent)
plt.imshow(output)
plt.show()
time_elapsed = (time.clock() - time_start)
print("Run time: %5.3f s" % time_elapsed)
```

```python
time_start=time.clock()
output=signal.convolve2d(high_pass,ascent)
plt.imshow(output)
plt.show()
time_elapsed = (time.clock() - time_start)
print("Run time: %5.3f s" % time_elapsed)
```
