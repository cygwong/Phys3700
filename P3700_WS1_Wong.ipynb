```python
import numpy as np
```

```python
ws = np.loadtxt("weatherstats_vancouver_daily.csv", delimiter = ",", usecols=(17),skiprows=(1))
print(ws)
```

```python
mws = np.mean(ws)
print("Mean of the Average Daily Wind Speed: %4.2f km/h" % (mws))
```

```python
import matplotlib.pyplot as plt
plt.xlabel("Wind Speed (km/h)"), plt.ylabel("Amount of data in range"), plt.title("Wind Speed Histogram")
plt.hist(ws,bins='auto')
plt.show()

```

```python
n, bins, patches = plt.hist(ws,75)
plt.xlabel("Wind Speed (km/h)"), plt.ylabel("Amount of data in range"), plt.title("Wind Speed Histogram")
plt.setp(patches[20], 'facecolor', 'g')
plt.setp(patches[14], 'facecolor', 'r')
plt.text(30, 600, r'$\bar{x}=$%5.2f  $\sigma=$%5.2f' %(np.mean(ws),np.std(ws)))
plt.axis([0,50,0,800])
plt.grid(True)
plt.show()
```

```python
a = np.random.randint(0,100,10000)
b = np.random.normal(0,100,10000)

am = np.mean(a) 
astd = np.std(a)
bm = np.mean(b)
bstd = np.std(b)

n, bins, patches = plt.hist(a,100)
plt.xlabel("numbers"), plt.ylabel("Amount of data in range"), plt.title("Randint")
plt.text(30, 130, r'$\bar{x}=$%5.2f  $\sigma=$%5.2f' %(am,astd))
plt.axis([0,100,0,150])
plt.grid(True)
plt.show()

n, bins, patches = plt.hist(b,100)
plt.xlabel("numbers"), plt.ylabel("Amount of data in range"), plt.title("normal")
plt.text(30, 350, r'$\bar{x}=$%5.2f  $\sigma=$%5.2f' %(bm,bstd))
plt.axis([-300,300,0,400])
plt.grid(True)

plt.show()
```
