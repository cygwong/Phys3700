```python
import numpy as np
import matplotlib.pyplot as plt
x= np.linspace(0,np.pi,1000)
H = 0.2*(1 + np.exp(-1j*x) + np.exp(-2j*x) + np.exp(-3j*x) +np.exp(-4j*x))
plt.plot(x,H)
plt.show()
```
