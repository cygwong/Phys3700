```python
import numpy as np
import matplotlib.pyplot as plt

A=2 # Amplitude of the sin wave
f=5.0 # frequency of the sin wave in Hz
T=1/f
fs=100.0*f
Ts=1/fs # horizontal spacing of the points that "fake" a continuous signal

fsamp=30
Tsamp=1/fsamp

cycles=50 # how many cycles of the wave to plot
tsamp=np.arange(0,cycles*T,Tsamp) # horizontal coordinate of your sin curve
xsamp=A*np.sin(2*np.pi*f*tsamp) # vertical value of your sin curve

t=np.arange(0,cycles*T,Ts) # horizontal coordinate of your sin curve
x=A*np.sin(2*np.pi*f*t) # vertical value of your sin curve
plt.figure(figsize=(12,4))
plt.axis([0,2,-2,2])
plt.plot(t,x)
plt.stem(tsamp,xsamp)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title("sin Curve")
plt.show()

fft = np.fft.fft(xsamp)
T = Tsamp  # sample rate
N = xsamp.size

# 1/T = frequency
f = np.linspace(0, 1 / T, N)
plt.figure(figsize=(12,4))
plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.stem(f,np.abs(fft)/N)
plt.show()
```
