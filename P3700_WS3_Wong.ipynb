```python
#1 a)

import numpy as np
import matplotlib.pyplot as plt

A=2  # Amplitude of the sin wave
f=5.0 # frequency of the sin wave in Hz
T=1/f
fs=100.0*f

Ts=1/fs # horizontal spacing of the points that "fake" a continuous signal
cycles=50  # how many cycles of the wave to save
t=np.arange(0,cycles*T,Ts) # horizontal coordinate of your sin curve
x=A*np.sin(2*np.pi*f*t) # vertical value of your sin curve

plt.figure(figsize=(12,4))
plt.axis([0,2,-2,2]) # we’re not plotting all cycles to make it easier to see what’s going on
plt.plot(t,x)
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.title("sin Curve")

A=2  # Amplitude of the sin wave
f=5.0 # frequency of the sin wave in Hz
T=1/f
fs=30.0#sampling frequency

Ts=1/fs # horizontal spacing of the points that "fake" a continuous signal
cycles=50  # how many cycles of the wave to save
t=np.arange(0,cycles*T,Ts) # horizontal coordinate of your sin curve
x=A*np.sin(2*np.pi*f*t) # vertical value of your sin curve

plt.stem(t,x)
plt.show()
```

```python
#2.a)
import numpy as np
import matplotlib.pyplot as plt


A=2  # Amplitude of the sin wave
f=5.0 # frequency of the sin wave in Hz
T=1/f
fs=100.0*f

Ts=1/fs # horizontal spacing of the points that "fake" a continuous signal
cycles=50  # how many cycles of the wave to save
t=np.arange(0,cycles*T,Ts) # horizontal coordinate of your sin curve
x=A*np.sin(2*np.pi*f*t) # vertical value of your sin curve

plt.figure(figsize=(12,4))
plt.axis([0,2,-2,2]) # we’re not plotting all cycles to make it easier to see what’s going on
plt.plot(t,x)
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.title("sin Curve")

A=2  # Amplitude of the sin wave
f=5.0 # frequency of the sin wave in Hz
T=1/f
fs=30.0#sampling frequency

Ts=1/fs # horizontal spacing of the points that "fake" a continuous signal
cycles=50  # how many cycles of the wave to save
t=np.arange(0,cycles*T,Ts) # horizontal coordinate of your sin curve
x=A*np.sin(2*np.pi*f*t) # vertical value of your sin curve

plt.stem(t,x)
plt.show()

fft = np.fft.fft(x) #xsamp is your stem plot data
N = x.size
f = np.linspace(0, fs, N) #Tsamp is the rate you sampled at (30 Hz, 12 Hz, 10 Hz, 7 Hz)
plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.stem(f,np.abs(fft)/N)
plt.show()
```
