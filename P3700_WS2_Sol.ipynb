```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,9)
x_fine=np.arange(0,9,0.01)
plt.figure(figsize=(8,8))
plt.axis([0,8,0,8])
plt.xlabel("Analog Sample Value")
plt.ylabel("Quantized Level")
plt.plot(x)
plt.step(x,x-1)
plt.show()

x_step=[0]

for n in range(99):
    x_step.append(0)
for m in range(8):
    for n in range(100):
        x_step.append(m+1)
        
#plt.axis([0,9,0,9])
plt.figure(figsize=(8,4))
plt.xlabel("Analog Sample Value")
plt.ylabel("Quantization Error")
plt.plot(x_fine,x_step-x_fine)
plt.show()
```

## #3

```python
x=np.arange(0,10)
x_fine=np.arange(0,10,0.01)

plt.figure(figsize=(8,8))
plt.axis([0,9,0,9])
plt.xlabel("Analog Sample Value")
plt.ylabel("Quantized Level")
plt.plot(x)
plt.step(x+.5,x)
plt.show()

x_step=[0]

for n in range(49):
    x_step.append(0)
for m in range(9):
    for n in range(100):
        x_step.append(m+1)
for n in range(50):
    x_step.append(10)

plt.figure(figsize=(8,4))
plt.xlabel("Analog Sample Value")
plt.ylabel("Quantization Error")
plt.plot(x_fine,x_step-x_fine)
plt.show()
```
