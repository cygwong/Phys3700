```python
#1 moving average filter
import numpy as np 
import matplotlib.pyplot as plt
wave =[0]*500
i = 0
while i < 500:
    if (i >= 200) and (i <= 299):
        wave[i] = 1
    i+=1
plt.plot(wave)
```

```python
#1.b)
noise = np.random.uniform(-0.6,0.6,500)
wave1 =[0]*500
i = 0
while i < 500:
    wave1[i]=noise[i]+wave[i]
    i+=1
plt.plot(wave1)
plt.show()
fft = np.fft.fft(wave1)
plt.stem(np.abs(fft))
plt.show()
```

```python
#1.c average the signal ignoring edge cases
y = [0]*500

i = 0
while i < 500:
    j = 0
    ave =0
    while j < 11:
        if(i+j < 500): #ignore end cases
            ave = ave + wave1[i+j]
        j+=1
    y[i] = ave/11
    i+=1
plt.plot(y)
plt.show()

fft = np.fft.fft(y)
plt.stem(np.abs(fft))
plt.show()
```
